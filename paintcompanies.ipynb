```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.coatingsworld.com/issues/2024-07-01/view_top-companies-report/the-2024-top-companies-report/"
```

```python
# Send an HTTP GET request to the website
response = requests.get(URL)

# Check if the request was successful
if response.status_code == 200:
    print(f"Successfully accessed {URL}")
else:
    print(f"Failed to access {URL} - Status code: {response.status_code}")
    exit()
```

```python
# Parse the website content with BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")
```

```python
url = 'https://www.coatingsworld.com/issues/2023-07-01/view_top-companies-report/top-companies-report-651116/'
```

```python
resp = requests.get(url)
page = BeautifulSoup(resp.text,"html.parser")
```

```python
list_com = page.find('div',id = "showInTextAds").find_all('a')[22:]
com_det=[]
for k in list_com:
  try:
    comp_index="".join(re.findall('[0-9]',k.text.strip()[0:2]))
    comp_name = "".join(re.findall('[a-zA-Z-]+',k.text.strip()))
    comp_rev ="".join(re.findall('[0-9]',k.text.strip()[2:]))
  except:
    comp_name='NaN'
    comp_rev="NaN"
    comp_index ="NaN"
  comp = {

      "Company Name": comp_name.replace('-',""),
      "Revenue_2023(USD)": comp_rev,
      "Position_in_2023": comp_index}
  com_det.append(comp)

```

```python
df1 = pd.DataFrame(com_det)
df1.head()
print(df1['Company Name'].unique())
```

```python
df2 = df.merge(df1,how='left',on='Company Name')
```

```python
df2=df2.iloc[:,[0,1,3,4,5,6,2,7,8]]
```
